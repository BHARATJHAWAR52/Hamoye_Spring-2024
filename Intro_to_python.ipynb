```python
import pandas as pd
url="https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
fuel=pd.read_csv(url, on_bad_lines='skip')
fuel.head()
```

```python
fuel.describe(include='all')
```

```python
#check for missing values
fuel.isnull().sum()
```

```python

#use groupby to count the sum of each unique value in the fuel unit column
fuel.groupby('fuel_unit')['fuel_unit'].count()
```

```python
fuel[['fuel_unit']] = fuel[['fuel_unit']].fillna(value='mcf')
#check if missing values have been filled
fuel.isnull().sum()
```

```python
fuel.groupby('report_year')['report_year'].count()
```

```python
#group by the fuel type code year and print the first entries in all the groups formed
fuel.groupby('fuel_type_code_pudl').first()
```

```python
fuel_df1 = fuel.iloc[0:19000].reset_index(drop=True)
fuel_df2 = fuel.iloc[19000:].reset_index(drop=True)
```

```python
#check that the length of both dataframes sum to the expected length
assert len(fuel) == (len(fuel_df1) + len(fuel_df2))
```

```python
#an inner merge will lose rows that do not match in both dataframes
pd.merge(fuel_df1, fuel_df2, how="inner")
```

```python
#outer merge returns all rows in both dataframes
pd.merge(fuel_df1, fuel_df2, how="outer")
```

```python
#removes rows from the right dataframe that do not have a match with the left and keeps all rows from the left
pd.merge(fuel_df1, fuel_df2, how="left")
```

```python
#check for duplicate rows
fuel.duplicated().any()
```

```python
# Import plotting library
import matplotlib.pyplot as plt
import seaborn as sns
fuel_unit = pd.DataFrame({'unit':['BBL', 'GAL', 'GRAMSU', 'KGU', 'MCF', 'MMBTU','MWDTH', 'MWHTH', 'TON'],'count':[7998, 84, 464, 110, 11354, 180, 95, 100, 8958]})

plt.figure(figsize=(7,4))
plt.xticks(rotation=90)
g = sns.barplot(data=fuel_unit, x='unit', y='count')
for i, bar in enumerate(g.patches):
    bar.set_color(sns.color_palette("husl", len(fuel_unit))[i])

g.set_yscale("log")
g.set_ylim(1, 12000)
plt.xlabel('Fuel Unit')
plt.show()
```

```python
# Select a sample of the dataset
sample_df = fuel.sample(n=50, random_state=4)
sns.regplot(x=sample_df["utility_id_ferc1"], y=sample_df["fuel_cost_per_mmbtu"],
fit_reg=False)
```

```python
# Box plot
sns.boxplot(x="fuel_type_code_pudl", y="utility_id_ferc1",
palette=["m", "g"], data=fuel)
```

```python
# KDE plot
sns.kdeplot(sample_df['fuel_cost_per_unit_burned'], shade=True, color="b")
```
