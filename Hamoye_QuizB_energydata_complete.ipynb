```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
energy_data=pd.read_csv("/content/drive/MyDrive/HAMOYE SPRING'24/energydata_complete (1).csv")
energy_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X = energy_data[['T2']]
y = energy_data['T6']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

model = LinearRegression()
model.fit(X_train, y_train)

# Making predictions on the test set
y_pred = model.predict(X_test)

# Calculating the RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Root Mean Squared Error:", round(rmse, 3))
```

```python
energy_data=energy_data.drop(columns=['date','lights'])
energy_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data= pd.DataFrame(scaler.fit_transform(energy_data),columns=energy_data.columns)
normalised_data
```

```python
from sklearn.metrics import mean_absolute_error
feature_df=energy_data.drop(columns=['Appliances'])

Appliances=energy_data['Appliances']

x_train, x_test, y_train, y_test = train_test_split(feature_df,Appliances,test_size=0.3,random_state=42)
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
y_train_predict=linear_model.predict(x_train)
mae=mean_absolute_error(y_train,y_train_predict)
round(mae,3)
```

```python
rmse = np.sqrt(mean_squared_error(y_train,y_train_predict))
round(rmse,3)
```

```python
y_test_predict=linear_model.predict(x_test)
mae=mean_absolute_error(y_test,y_test_predict)
round(mae,3)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_test_predict))
round(rmse,3)
```

in the training dataset the model has 53.747 mae and 95.216 rmse whereas in the test dataset the model has 53.647 mae and 93.64 rmse. Since the model doesnot show much of a difference in the training and test dataset on the basis of mae and rmse metrices we can say the model does not overfit.


```python
from sklearn.linear_model import Ridge

ridge_model = Ridge()
ridge_model.fit(x_train, y_train)

y_test_pred_ridge = ridge_model.predict(x_test)
rmse_test_ridge = np.sqrt(mean_squared_error(y_test, y_test_pred_ridge))
round(rmse_test_ridge,3)
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso()
lasso_model.fit(x_test, y_test)

y_test_pred_lasso = lasso_model.predict(x_test)
rmse_test_lasso = np.sqrt(mean_squared_error(y_test, y_test_pred_ridge))
round(rmse_test_lasso,3)
```
